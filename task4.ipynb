```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Funktion zum Einlesen der Daten aus der Datei
def read_data(filename):
    data = np.genfromtxt(filename, delimiter='\t', skip_header=1)
    return data
```

```python
# Einlesen der Reset-Daten
reset_data = read_data("./data/Reset_switching_dynamics_3V.txt")
pulse_time = reset_data[:, 0]
resistance = reset_data[:, 2] / reset_data[:, 1]

# Plot der Reset-Daten
plt.figure(figsize=(10, 5))
plt.plot(pulse_time, resistance, label='Reset @ 3V', marker='o')
plt.title('Reset Switching Dynamics @ 3V')
plt.xlabel('Pulse Time (s)')
plt.ylabel('Resistance ' + r'$\Omega$')
plt.legend()
plt.xscale('log')
plt.ylim(0)
plt.grid(True)
plt.show()
```

```python
# Einlesen der Set-Daten
set_data = read_data("./data/Set_switching_dynamics_-1_75V.txt")
pulse_time = reset_data[:, 0]
resistance = reset_data[:, 2] / reset_data[:, 1]

# Plot der Set-Daten
plt.figure(figsize=(10, 5))
plt.plot(set_data[:, 0], set_data[:, 2] / set_data[:, 1], label='Set @ -1.75V', marker='o')
plt.title('Set Switching Dynamics @ -1.75V')
plt.xlabel('Pulse Time (s)')
plt.ylabel('Resistance ' + r'$\Omega$')
plt.legend()
plt.xscale('log')
plt.ylim(0)
plt.grid(True)
plt.show()
```
